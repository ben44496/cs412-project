```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../job-postings-headline-index.csv")
df = df[['dateString', 'countryName', 'value']]
df['dateString'] = pd.to_datetime(df['dateString'])
df
```

```python
# Rename columns to "Date", "Country", "Value"
# Drop index
df = df.rename(columns={'dateString': 'Date', 'countryName': 'Country', 'value': 'Value'})
df = df.reset_index(drop=True)
df.head()
```

```python
df.to_csv("../clean/indeed_job_index.csv", index=False)
```

```python
import plotly.express as px
fig = px.line(df, x='Date', y='Value', color='Country', title='Indeed Job Index')
fig.show()
```

```python
# Plot log returns
log_returns = np.log(df['Value']).diff()
fig = px.line(log_returns, title='Indeed Job Index Log Returns')
fig.show()
```
